#### Get unplanned absences

```python
dfs = {}
for i, sheet in enumerate(sheet_names):
    df = pd.read_excel(full_path, sheet, usecols="S,T,AD,AE,AO,AP,AZ,BA")
    df.columns = ['CPK 1', 'ICT 1', 'CPK 2', 'ICT 2', 'CPK 3', 'ICT 3', 'CPK 4', 'ICT 4']
    df = df.iloc[pd.np.r_[4:7, 14:19, 26:29]].reset_index(drop=True)
    df.index = ['Trim 1', 'Trim 2', 'Door Line', 'Engine Line', 'Chassis 1 & 2',
                'Sunroof Deck', 'Chassis 3', 'Chassis 4', 'Final 1', 'Final 2', 'Rolls']
    df = df.abs()
    df['dept'] = ''
    df['dept'].iloc[0:3] = 'Trim'
    df['dept'].iloc[3:8] = 'Chassis'
    df['dept'].iloc[8:11] = 'Final'
    key = 'Crew ' + chr(i + 65)
    dfs[key] = df
dfs
```

```python
df_all = pd.concat(dfs).reset_index()
df_all = df_all.rename(columns={'level_0': 'crew', 'level_1': 'production_line'})
df_all
```

```python
group_cols = ['crew', 'dept', 'production_line']
group_cols
```

#### ICT Calculations

```python
ict_cols = [col for col in df_all.columns if 'ICT' in col]
ict_cols
```

```python
df_ict = df_all[group_cols + ict_cols]
df_ict
```

```python
df_ict_vals = pd.melt(df_ict, id_vars=group_cols, value_vars=ict_cols)
df_ict_vals
```

```python
df_ict_vals['value'] = df_ict_vals['value'].astype(int)
```

```python
df_ict_mae = df_ict_vals.groupby(['crew', 'dept'])['value'].mean().reset_index()
df_ict_mae
```

```python
df_ict_vals['value'].mean()
```
